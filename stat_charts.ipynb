```python
import matplotlib.pyplot as plt
import helper
```

```python
stats = helper.unpickle("save_stats.json")

episodes = [stats["ep_step"] *(i+1) for i in range(len(stats["diamonds"]))]
```

```python
plt.plot(episodes, stats["best_policies"])

plt.title("Best Policy")
plt.xlabel("Episodes of training")
plt.ylabel("Cumulative Reward")
```

```python
plt.plot(episodes, stats["diamonds"])

plt.xlabel("Episodes of training")
plt.ylabel("Diamonds Found")
```

```python
plt.plot(episodes, stats["ores"])

plt.xlabel("Episodes of training")
plt.ylabel("Ores Found")
```
